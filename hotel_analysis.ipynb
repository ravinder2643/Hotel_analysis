*Data Analysis and Visulatization*

```python
# calculate cancell percebtage of hotels

cancelled_per= df['is_canceled'].value_counts(normalize= True)
```

```python
cancelled_per
```

```python
cancelled_per= df['is_canceled'].value_counts(normalize= True)
print(cancelled_per)

plt.figure(figsize= (5,4))
plt.title("Reservation status count")
plt.bar(['Not_canceled', 'Canceled'],df['is_canceled'].value_counts() ,edgecolor='k', width=0.7) 
```

```python
#depending on hotel whoses cancelation ratio is greater


plt.figure(figsize= (5,4))
ax1= sns.countplot(x= 'hotel' , hue= 'is_canceled', data = df)
plt.title("Reservation status in different hotels")
plt.xlabel('hotel')
plt.ylabel('numner_of_reservations')
```

```python
#resort hotels me kitna %cancell ho rh ahia kitna nhi 

resort_hotel= df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize= True)
```

```python
# same for city hotels
city_hotel= df[df['hotel'] == "City Hotel"]
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
# lets check  is there any effet of price on resolt and city hotels cancellation
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()


resort_hotel
city_hotel
```

```python
# now visualize it 

plt.figure(figsize= (20,8))
plt.title('Average daily rate un city and resort hotel',  fontsize= 30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label= 'resort_hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label= 'city_hotel')
plt.legend(fontsize= 30, loc= 'best')
```

```python
#in which month most reservation  happend and which months we have more cancelation

df['month']= df['reservation_status_date'].dt.month

plt.figure(figsize= (16,8))
ax1= sns.countplot(x= 'month', hue= 'is_canceled', data= df)
plt.title('Reservation status per month', fontsize= 30)
plt.xlabel("months")
plt.ylabel('number of resservations')
plt.legend(['is_canceled','canceled'])
```

```python
#checking the effect of price on cancellation rate

plt.figure(figsize= (16,8))
plt.title("ADR per month", fontsize=30)
sns.barplot(x='month', y='adr', data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
```

```python
#this proves more price leads to MORE Cncelation
```

```python
#lets check hich country having more cancelation

canceled_data= df[df['is_canceled']==1]
top_10_country= canceled_data['country'].value_counts().head(10)
plt.figure(figsize=(8,8))
plt.title("Top 10 Countries with reservation cancelled")
plt.pie(top_10_country, autopct= '%.2f', labels= top_10_country.index)
plt.show()
```

```python
#lets see where the clients are comming from either from online , travel agents etc

df['market_segment'].value_counts()
```

```python
df['market_segment'].value_counts(normalize= True)
```

```python
# same thing in cancelled data
canceled_data['market_segment'].value_counts(normalize= True)
```

```python
cancelled_df_adr= canceled_data.groupby("reservation_status_date")[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date', inplace= True)

not_canceled_data= df[df['is_canceled']==0]
not_cancelled_df_adr= not_canceled_data.groupby("reservation_status_date")[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace= True)

plt.figure(figsize= (20,6))
plt.title("Average daily rate")
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'], label='not_cancelled_df_adr')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='cancelled_df_adr')
plt.legend()
```
